```python
#Load Libraries
import pandas as pd
import numpy as np
from imblearn.over_sampling import SMOTE
```

```python
train_df=pd.read_csv('imputed_training_data.csv')
columns=list(train_df.columns)
```

```python
%%time
xTrain=train_df.drop(['STAT_CAUSE_DESCR'], axis=1).values
yTrain=train_df['STAT_CAUSE_DESCR']

oversample = SMOTE()
xTrain, yTrain = oversample.fit_resample(xTrain, yTrain)

dfx = pd.DataFrame(yTrain)
dfx.value_counts().plot.bar()
```

```python
smote_df=pd.DataFrame(xTrain,columns=columns[0:(len(columns)-1)])
smote_df['STAT_CAUSE_DESCR']=yTrain
```

```python
%%time
smote_df.to_csv('SMOTE_fires_data.csv', index=False)
```
